```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_tobs = session.query(Measurement.tobs).\
    filter(func.strftime("%m", Measurement.date) == "06").all()
june_tobs
```

```python
#  3. Convert the June temperatures to a list.
junetobs_list = list(np.ravel(june_tobs))
junetobs_list
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
junetobs_df = pd.DataFrame(junetobs_list)
junetobs_df.rename(columns={0: 'June Temps'}, inplace=True)
junetobs_df.head()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
junetobs_df.describe()
```

## D3: Two additional queries that you would perform to gather more weather data for June and December.

```python
### mean temperature per year to understand if there are large year to year changes.  Example done with Dec, but the code could be easily converted to do the same anaalysi for June
```

```python
#setting the query and creating the dataframe
dec_tobs_year = session.query(Measurement.tobs, Measurement.date).\
    filter(func.strftime("%m", Measurement.date) == "12").all()
dec_tobs_year_df = pd.DataFrame(dec_tobs_year)
dec_tobs_year_df.rename(columns={0: 'tobs', 1: 'date'}, inplace=True)

#converting the date from object to datetime datatype
dec_tobs_year_df["date"] = pd.to_datetime(dec_tobs_year_df["date"])
#add a year column to make yearly grouping easier
dec_tobs_year_df['year']=dec_tobs_year_df["date"].dt.year
#get the daily averages
dec_daily_avg = dec_tobs_year_df.groupby(dec_tobs_year_df["date"]).mean()
#get the yearly averages
dec_daily_avg.groupby(dec_daily_avg["year"]).mean()
```

```python
### number of days in Dec with either precipitation >=0.5 inches or temp below 65 degrees as either of these conditions would presumably lead to lower sales
```

```python
#setting the query and creating the dataframe
dec_prcp_and_tobs_year = session.query(Measurement.tobs, Measurement.prcp, Measurement.date).\
    filter(func.strftime("%m", Measurement.date) == "12").filter((Measurement.tobs < "65") | (Measurement.prcp >= "0.5")).all()
dec_prcp_and_tobs_year_df = pd.DataFrame(dec_prcp_and_tobs_year)
dec_prcp_and_tobs_year_df.rename(columns={0: 'tobs', 1: 'prcp', 2: 'date'}, inplace=True)
#convert date from object to datetime datatype
dec_prcp_and_tobs_year_df["date"] = pd.to_datetime(dec_prcp_and_tobs_year_df["date"])
#get the daily averages
crummy_days = dec_prcp_and_tobs_year_df.groupby(dec_prcp_and_tobs_year_df["date"]).mean()
#reset index and add a year column
crummy_days=crummy_days.reset_index()
crummy_days['yr']=crummy_days['date'].dt.year
#get the count per year
crummy_days.groupby(crummy_days["yr"]).count()["date"]
```
